```python
# Python Libary that allows us to use the feature to display a photo
from IPython.display import Image
```

```python
# We, the programmers, have to provide Zumi with some address/location of where the picture we want to upload 
# exists.  Instead of providing the end location, such as 'demoPic.jpg' we need to provide the whole path.

my_profile_name =  'Thomas' # TODO:: edit this line of code, everyones will be different unless your name is Thomas


root_path = '/home/pi/Dashboard/user/'
me_path   = my_profile_name + '/MyImages/'
fileName  = 'demoPic.jpg'
```

```python
display( Image(filename=(root_path+me_path+fileName)) )
```

```python
# changing the size of the image that we import
display( Image(filename=(root_path+me_path+fileName), width=300, height=200) )
```
